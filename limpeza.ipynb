```python
import pandas as pd

# Read the downloaded Excel file with specified engine
df = pd.read_excel('mapeamento_estudantes.xlsx')

# Display the DataFrame
print(df.info)
```

```python
# Define the column name
column_name = 'Qual é o seu número de matrícula?'

# Clean the data in the specified column
df[column_name] = df[column_name].astype(str)  # Ensure the column is treated as strings
df[column_name] = df[column_name].str.replace(' ', '')  # Remove white spaces
df[column_name] = df[column_name].str.replace('-', '')  # Remove hyphens

# Filter numbers with length between 8 and 9
df = df[df[column_name].str.len().between(8, 9)]

# Drop duplicates
df = df.drop_duplicates(subset=[column_name])

# Display the cleaned DataFrame
print(df.info)
```
